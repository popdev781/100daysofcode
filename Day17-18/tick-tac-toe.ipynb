```python
#! /usr/bin/python

def init_board():
    new_board = [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ]
    return new_board

def display_board(board):
    for i in range(3):
        for j in range(3):
            print(board[i][j], end=" ")
        print()
        
def ask_move():
    position = input("Enter the position where you want to move to ")
    # print("you want to move :"+ position)
    return position

def set_x_y(position):
    pos = int(position)
    x,y=0,0
    if pos <=3:
        x=0
        y=pos-1
    elif pos <=6:
        x = 1
        y = pos-4
    else:
        x = 2
        y = pos-7
    return (x,y)

def check_win(board, user):
    # 1st check if any row or column matches
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] == user:
            return True
        elif board[0][i] == board[1][i] == board[2][i] == user:
            return True
    
    # 2nd check if any cross match
    if board[0][0] == board[1][1] == board[2][2] == user:
        return True
    elif board[0][2] == board[1][1] == board[2][0] == user:
        return True
    else:
        return False

def update_board(board, position, user):
    pos = position
    temp_board= board
    user = user
    x,y = set_x_y(pos)
    while (str(temp_board[x][y]) in "xo"):
        print("position is alreay occupied, try again.")
        display_board(board)
        pos = ask_move()
        x,y = set_x_y(pos)
    else:
        temp_board[x][y] = user;
              
    return temp_board
```

```python
ans = input("Are you ready to play(y/n):")
while ans=='y' or ans == "Y":
    board = init_board()
    print("we are ready to play.")
    user1 = input("select player(O/X)").lower()
    if user1 != "x"  and user1 != "y":
        print("you should have selected \"o\" or \"x\". Exiting.")
        break;              
              
    user2 = "x" if user1=="o" else "o"
    print("")
    display_board(board)
    #print("user1 is ",user1)
    #print("user2 is ", user2)
    for i in range(1, 10):
        current_user = user1 if i%2==1 else user2
        position = ask_move()
        #print(position)
        while True:
            if not position.isdigit() or int(position) >9 or int(position) <1:
                print("wrong input for position, it can only be between 1 and 9. try again.")
                position = ask_move()
            else: break;
            
        move=i+1
        board = update_board(board, position, current_user)
        display_board(board)
        if move == "9":
            if check_win(board, current_user):
                print("user", current_user, " win.")
                break;
            else:
                print("Draw")
        elif move>=6 :
            if check_win(board, current_user):
                print("user", current_user, " win.")
                break;
            else:
                pass
    ans = input("Game finished. wanna try again? (y|n)")
    
```
